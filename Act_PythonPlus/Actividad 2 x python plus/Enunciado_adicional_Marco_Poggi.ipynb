*   dado alguno de los datasets elegidos para el trabajo integrador, mostrar una nube de palabras (con la librería wordcloud) con las palabras elegidas para el juego de acuerdo a alguno de los criterios elegidos en su trabajo.  Considerar utilizar la opción de que se puedan repetir las palabras.

```python
import pandas as pd
from wordcloud import WordCloud, wordcloud
import matplotlib.pyplot as plt

def lista_palabras():
    archivo = 'Pokemon.csv'
    df = pd.read_csv(archivo,encoding='UTF-8')
    new_df = df[(df["Type 1"] == "Fire") | (df["Type 1"] == "Flying")]
    return new_df['Name']

text = str(lista_palabras())

wordcloud = WordCloud(width = 800, height = 600,background_color='white', colormap='RdPu', collocations=True).generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

Criterio utilizado: se obtiene los nombres de los pokemones que son de tipo fuego o volador
