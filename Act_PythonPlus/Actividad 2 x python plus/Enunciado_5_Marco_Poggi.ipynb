* *dado el archivo de registros de jugadas (eventos) de MemPy, mostrar un gráfico con los usuarios que más veces jugaron el juego (independientemente del resultado de la partida.*



````python
#Solucion
import pandas as pd
from matplotlib import pyplot as plt

def get_jugadores(data):
    data = data[data['Nombre de evento'] == 'fin']
    nombres = data['Usuarie - nick'].value_counts()
    return nombres

def graficar(etiquetas,jugadores):
    plt.pie(jugadores,labels = etiquetas,autopct="%0.1f %%")
    plt.axis('equal')
    plt.show()

def main():
    data = pd.read_csv('datos de prueba.csv')
    jugadores = get_jugadores(data)
    graficar(jugadores.keys(),jugadores) 
main()
````

Lo primero es importar las librerias con las que se va a trabajar, en este caso ***pandas*** y ***matplotlib***.

````python
import pandas as pd
from matplotlib import pyplot as plt
````

Lo siguiente que se observa en el codigo es que se llama a la funcion ***main()***

````python
def main():
    data = pd.read_csv('datos de prueba.csv')
    jugadores = get_jugadores(data)
    graficar(jugadores.keys(),jugadores) 
````

Lo primero que hace es guardar en la variable `data`, un dataframe generado mediante pandas y su funcion `read_csv()`.


*   `data = pd.read_csv('datos de prueba.csv')`.

Luego en la variable `jugadores` se almacena lo retornado por la funcion `get_jugadores(data)`.

Y por ultimo se llama a la funcion `graficar()`, que recibe como parametro las etiquetas y los juegadores.




````python
def get_jugadores(data):
    data = data[data['Nombre de evento'] == 'fin']
    nombres = data['Usuarie - nick'].value_counts()
    return nombres
````

La funcion get_jugadores(), lo que hace es primero en la variable data guardar una especie de filtro que almacenara un nuevo dataframe con las filas que unicamente tiene como nombre de evento 'fin'


*   `data = data[data['Nombre de evento'] == 'fin']`

Despues en nombres se obtiene un elemento de tipo 'pandas.core.series.Series'
y mediante la funcion value_counts() que tiene este tipo de elemento, podemos obtener cuantas apariciones hay de cada nombre.

*   `nombres = data['Usuarie - nick'].value_counts()`

y retornamos dicha variable


*   `return nombres`

Es decir que en jugadores vamos a tener este elemento de tipo series

`jugadores = get_jugadores(data)`

que si hacemos print(jugadores) devuelve algo asi:
```
jose          2  
clara         1  
raul          1  
flor          1

Name: Usuarie - nick, dtype: int64
```










````python
def graficar(etiquetas,jugadores):
    plt.pie(jugadores,labels = etiquetas,autopct="%0.1f %%")
    plt.axis('equal')
    plt.show()
````

Por ultimo la funcion `graficar()` nos permite visualizar mediante **matplotlib** el grafico de los jugadores que mas jugaron al juego.  
Lo primero que hacemos es mediante `plt.pie()`, que nos permite crear un grafico estilo torta, pasar los `jugadores` es decir el elemento series que obtuvimos anteriormente, luego las labels o etiquetas que seran `['jose', 'raul', 'pepe', 'clementina', 'clara', 'flor']`, y con la propiedad `autopct`, lo que buscamos es mostrar un valor porcentual con formato string.

`plt.axis('equal')` lo utilizamos para que tango el eje x y el eje y tengan la misma longitud


Y por ultimo para mostrar el grafico utilizamos `plt.show()`
